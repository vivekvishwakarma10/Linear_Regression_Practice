```python
df = data.drop('car name' , axis = 1)
```

```python
df['origin'] = df['origin'].replace({1: 'america', 2: 'europe', 3 : 'asia'})
```

# we cant build model with this bcz values should be in integer or in float

```python
df= pd.get_dummies(df, columns= ['origin']) # converting categorial values to dummy values
```

```python
digid_in_hp =  pd.DataFrame(df.horsepower.str.isdigit())
```

# handling the missing values

```python
sns.distplot(df['horsepower'])
```

```python
df['horsepower'] =  df['horsepower'].fillna(df['horsepower'].median)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# converting hp column obj to float
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```
